```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Step 1: Preprocess the data
data = pd.read_csv('withoutloop.csv')
positions = data[['x_axis', 'y_axis']].values

# Step 2: Split the data
train_data, test_data = train_test_split(positions, test_size=0.2, shuffle=False)

# Step 3: Prepare the input and output sequences
sequence_length = 10  # Adjust as needed
X_train, y_train = [], []
for i in range(sequence_length, len(train_data)):
    X_train.append(train_data[i - sequence_length:i])
    y_train.append(train_data[i])

X_test, y_test = [], []
for i in range(sequence_length, len(test_data)):
    X_test.append(test_data[i - sequence_length:i])
    y_test.append(test_data[i])

# Step 4: Reshape and normalize the data
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)

# Normalize the data between 0 and 1
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, 2)).reshape(-1, sequence_length, 2)
X_test = scaler.transform(X_test.reshape(-1, 2)).reshape(-1, sequence_length, 2)
y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)

# Step 5: Build and train the LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(sequence_length, 2)))
model.add(Dense(2))
model.compile(loss='mse', optimizer='adam')
model.fit(X_train, y_train, epochs=50, batch_size=32)

# Step 6: Evaluate the model
loss = model.evaluate(X_test, y_test)
print('Test Loss:', loss)

# Step 7: Predict the enemy path
predictions = model.predict(X_test)
predicted_coordinates = scaler.inverse_transform(predictions)

# Plot the predicted path
plt.figure(figsize=(10, 6))
plt.plot(test_data[:, 0], test_data[:, 1], label='Actual Path')
plt.plot(predicted_coordinates[:, 0], predicted_coordinates[:, 1], label='Predicted Path')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Enemy Path Prediction')
plt.legend()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

def read_data(filename):
    points = []
    with open(filename, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        for row in csvreader:
            x, y = map(float, row)
            points.append((x, y))
    return points

def calculate_bezier_point(t, control_points):
    n = len(control_points) - 1
    point = np.zeros(2)
    for i in range(n + 1):
        coefficient = np.math.comb(n, i) * t**i * (1 - t)**(n - i)
        point += coefficient * np.array(control_points[i])
    return point

def predict_path(points, num_segments):
    predicted_points = []
    for i in range(0, len(points) - 3, 3):
        control_points = points[i:i+4]
        for j in range(num_segments + 1):  # Include the last point
            t = j / num_segments
            predicted_point = calculate_bezier_point(t, control_points)
            predicted_points.append(predicted_point)
    return predicted_points

def calculate_accuracy(actual_points, predicted_points):
    errors = []
    for actual, predicted in zip(actual_points, predicted_points):
        error = np.linalg.norm(np.array(actual) - np.array(predicted)) ** 2
        errors.append(error)
    mse = np.mean(errors)
    return mse

def plot_path(actual_points, predicted_points):
    actual_x, actual_y = zip(*actual_points)
    predicted_x, predicted_y = zip(*predicted_points)
    plt.plot(actual_x, actual_y, 'bo', label='Actual Path')
    plt.plot(predicted_x, predicted_y, 'r-', label='Predicted Path')
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.title('Path Prediction using Bezier Curve')
    plt.legend()
    plt.show()

# Read data from CSV file
data_points = read_data('withoutloop.csv')

# Normalize the data
scaler = MinMaxScaler()
data_points = scaler.fit_transform(data_points)

# Set the number of segments for the Bezier curve
num_segments = 100

# Predict the path using Bezier curve
predicted_path = predict_path(data_points, num_segments)

# Calculate the accuracy of the prediction
mse = calculate_accuracy(data_points, predicted_path)
print(f"Mean Squared Error: {mse}")

# Plot the actual and predicted paths
plot_path(data_points, predicted_path)
```
